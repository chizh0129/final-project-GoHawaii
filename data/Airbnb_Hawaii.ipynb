# Data Description

```
# describle listings_summary
listings_summary0[['price', 'availability_365']].describe().style.format({'price' : '{:.2f}', 'availability_365' : '{:.2f}'})
```

```
# group listings by neighbourhood_group
listingsBy0 = listings_summary0.groupby("neighbourhood_group")
```

```
listingsBy0.first()
```

```
listingsBy1 = listingsBy0.agg({'id':'size', 'price':['mean'], 'availability_365':['mean']}).reset_index()
listingsBy1.columns = ['County', 'Count_of_Listing', 'Mean_Price', 'Mean_Available_Days']
listingsBy1.style.format({'Mean_Price' : '{:.2f}', 'Mean_Available_Days' : '{:.2f}'})
```
